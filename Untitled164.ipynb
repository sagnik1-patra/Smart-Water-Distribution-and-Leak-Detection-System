```python
import os, json, yaml, pickle, random
import numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ---------------------------------------------------------------
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Water Distribution and Leak Detection System\archive\water_leak_detection_1000_rows.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Water Distribution and Leak Detection System\archive"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ---------------------------------------------------------------
# Load & preprocess
# ---------------------------------------------------------------
df = pd.read_csv(DATA_PATH)
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
df = df.dropna().reset_index(drop=True)

encoders = {}
for c in df.columns:
    if df[c].dtype == "object":
        le = LabelEncoder()
        df[c] = le.fit_transform(df[c].astype(str))
        encoders[c] = le

feature_cols = [c for c in df.columns if c not in ["timestamp","leak_status","burst_status"]]
target_col = "burst_status" if "burst_status" in df.columns else feature_cols[-1]

X, y = df[feature_cols].values, df[target_col].values
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

def make_seq(X,y,steps=5):
    Xs,ys=[],[]
    for i in range(len(X)-steps):
        Xs.append(X[i:i+steps]); ys.append(y[i+steps])
    return np.array(Xs),np.array(ys)

TIMESTEPS=5
X_seq,y_seq=make_seq(X_scaled,y,TIMESTEPS)
split=int(0.8*len(X_seq))
X_train,X_test=X_seq[:split],X_seq[split:]
y_train,y_test=y_seq[:split],y_seq[split:]

# ---------------------------------------------------------------
def build_lstm(params):
    h,d,lr=int(params[0]),float(params[1]),float(params[2])
    model=Sequential([
        LSTM(h,input_shape=(TIMESTEPS,X_train.shape[2])),
        Dropout(d),
        Dense(32,activation='relu'),
        Dense(1,activation='linear')
    ])
    model.compile(optimizer=Adam(learning_rate=lr),loss='mse')
    es=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
    hist=model.fit(X_train,y_train,validation_data=(X_test,y_test),
                   epochs=40,batch_size=16,verbose=0,callbacks=[es])
    return np.min(hist.history["val_loss"]),model,hist

# ---------------------------------------------------------------
# Hybrid AIS + QPSO
# ---------------------------------------------------------------
def qpso_optimize(pop_size=8,gens=5,beta=1.0):
    lb=np.array([32,0.1,0.0005]); ub=np.array([128,0.5,0.01])
    pos=np.random.uniform(lb,ub,(pop_size,3))
    pbest=pos.copy(); pbest_fit=np.full(pop_size,np.inf)
    gbest=None; gbest_fit=np.inf; loss_curve=[]

    for g in range(gens):
        print(f"üåê Generation {g+1}/{gens}")
        # ---- Evaluate ----
        new_fit=[]
        for i in range(len(pos)):
            val,model,hist=build_lstm(pos[i])
            if val<pbest_fit[i] if i<len(pbest_fit) else True:
                if i>=len(pbest_fit):
                    pbest_fit=np.append(pbest_fit,val)
                    pbest=np.vstack((pbest,pos[i]))
                else:
                    pbest[i],pbest_fit[i]=pos[i],val
            new_fit.append(val)
            if val<gbest_fit:
                gbest,gbest_fit,gbest_model,gbest_hist=pos[i].copy(),val,model,hist
        print(f"  üß¨ Best so far: {gbest.round(4).tolist()} ‚Üí ValLoss={gbest_fit:.4f}")

        # ---- AIS cloning ----
        top_idx=np.argsort(new_fit)[:3]
        clones=[]
        for idx in top_idx:
            for _ in range(2):
                c=pos[idx].copy()
                mut=(np.random.rand(3)<0.3)
                noise=np.random.uniform(-0.1,0.1,3)
                c=np.clip(c+mut*noise*(ub-lb),lb,ub)
                clones.append(c)
        pos=np.vstack((pos,clones))

        # ---- Quantum update ----
        mbest=np.mean(pbest,axis=0)
        new_pos=[]
        for i in range(len(pos)):
            u=np.random.rand(3)
            u[u==0]=1e-8  # avoid log(0)
            phi=beta*np.abs(mbest-pos[i])*np.log(1/u)
            p=random.random()
            qpso_vec=p*pbest[i%len(pbest)] + (1-p)*gbest + np.sign(np.random.rand(3)-0.5)*phi
            new_pos.append(np.clip(qpso_vec,lb,ub))
        pos=np.array(new_pos)

        # maintain equal length for pbest_fit
        if len(pbest_fit)<len(pos):
            add=len(pos)-len(pbest_fit)
            pbest_fit=np.concatenate([pbest_fit,np.full(add,np.inf)])
            pbest=np.vstack((pbest,pos[:add]))
        loss_curve.append(gbest_fit)
        print(f"  ‚úÖ Generation {g+1} best loss: {gbest_fit:.4f}\n")
    return gbest,gbest_model,gbest_hist,loss_curve

# ---------------------------------------------------------------
best_params,best_model,best_hist,loss_curve=qpso_optimize()

# ---------------------------------------------------------------
# Evaluate final model
# ---------------------------------------------------------------
y_pred=best_model.predict(X_test).flatten()
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
rmse=np.sqrt(mse)
print(f"[RESULT] RMSE={rmse:.4f}, MAE={mae:.4f}, R¬≤={r2:.4f}")

# ---------------------------------------------------------------
# Save artifacts
# ---------------------------------------------------------------
best_model.save(os.path.join(OUTPUT_DIR,"ais_qpso_model.h5"))
with open(os.path.join(OUTPUT_DIR,"scaler.pkl"),"wb") as f: pickle.dump(scaler,f)
with open(os.path.join(OUTPUT_DIR,"encoders.pkl"),"wb") as f: pickle.dump(encoders,f)
json.dump({
    "RMSE":float(rmse),"MAE":float(mae),"R2":float(r2),
    "Best_Params":best_params.tolist(),"Loss_Curve":[float(x) for x in loss_curve]
},open(os.path.join(OUTPUT_DIR,"ais_qpso_results.json"),"w"),indent=4)

# ---------------------------------------------------------------
# Graphs
# ---------------------------------------------------------------
plt.figure(figsize=(10,5))
plt.plot(y_test[:150],label="Actual",marker='o')
plt.plot(y_pred[:150],label="Predicted",marker='x')
plt.title("AIS + QPSO Prediction vs Actual")
plt.xlabel("Sample Index"); plt.ylabel("Leak/Burst Index")
plt.legend(); plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR,"ais_qpso_prediction_graph.png"))
plt.show()

plt.figure(figsize=(8,5))
plt.plot(loss_curve,marker='o')
plt.title("AIS + QPSO Optimization Loss per Generation")
plt.xlabel("Generation"); plt.ylabel("Best Validation Loss")
plt.grid(True); plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR,"ais_qpso_optimization_loss.png"))
plt.show()

print("\n‚úÖ AIS + QPSO (Stable v2) completed successfully!")
print("üìÅ Results saved in:", OUTPUT_DIR)
print("üìà Main graph: ais_qpso_prediction_graph.png")
```
