```python
import os
import json
import yaml
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import pickle
import tensorflow as tf

# ---------------------------------------------------------------
# 1️⃣ Configuration
# ---------------------------------------------------------------
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Water Distribution and Leak Detection System\archive\water_leak_detection_1000_rows.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Water Distribution and Leak Detection System\archive"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ---------------------------------------------------------------
# 2️⃣ Load Dataset
# ---------------------------------------------------------------
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA_PATH)

print("[INFO] Shape:", df.shape)
print("[INFO] Columns:", df.columns.tolist())

# Clean column names
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

# Drop missing values
df = df.dropna().reset_index(drop=True)

# ---------------------------------------------------------------
# 3️⃣ Encode Non-Numeric Columns
# ---------------------------------------------------------------
encoders = {}
for col in df.columns:
    if df[col].dtype == "object":
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))
        encoders[col] = le
        print(f"[ENCODE] Column '{col}' encoded with {len(le.classes_)} unique values.")

# ---------------------------------------------------------------
# 4️⃣ Feature Selection
# ---------------------------------------------------------------
feature_cols = [col for col in df.columns if col not in ["timestamp", "leak_status", "burst_status"]]
target_col = "burst_status" if "burst_status" in df.columns else feature_cols[-1]

print(f"[INFO] Using {len(feature_cols)} features → {feature_cols}")
print(f"[INFO] Target column → {target_col}")

X = df[feature_cols].values
y = df[target_col].values

# ---------------------------------------------------------------
# 5️⃣ Scale Data
# ---------------------------------------------------------------
scaler_x = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)

# Save scaler
scaler_path = os.path.join(OUTPUT_DIR, "scaler.pkl")
with open(scaler_path, "wb") as f:
    pickle.dump(scaler_x, f)
print(f"[INFO] Scaler saved at: {scaler_path}")

# Save label encoders (for reproducibility)
enc_path = os.path.join(OUTPUT_DIR, "encoders.pkl")
with open(enc_path, "wb") as f:
    pickle.dump(encoders, f)
print(f"[INFO] Encoders saved at: {enc_path}")

# ---------------------------------------------------------------
# 6️⃣ Sequence Preparation for LSTM
# ---------------------------------------------------------------
def create_sequences(X, y, timesteps=5):
    Xs, ys = [], []
    for i in range(len(X) - timesteps):
        Xs.append(X[i:(i + timesteps)])
        ys.append(y[i + timesteps])
    return np.array(Xs), np.array(ys)

TIMESTEPS = 5
X_seq, y_seq = create_sequences(X_scaled, y, TIMESTEPS)

split = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]

print(f"[INFO] Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# ---------------------------------------------------------------
# 7️⃣ Build LSTM Model
# ---------------------------------------------------------------
model = Sequential([
    LSTM(64, input_shape=(TIMESTEPS, X_train.shape[2]), return_sequences=False),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()

# ---------------------------------------------------------------
# 8️⃣ Train
# ---------------------------------------------------------------
es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=16,
    callbacks=[es],
    verbose=1
)

# ---------------------------------------------------------------
# 9️⃣ Evaluate
# ---------------------------------------------------------------
y_pred = model.predict(X_test).flatten()

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"[RESULT] MSE: {mse:.4f}")
print(f"[RESULT] R² Score: {r2:.4f}")

# ---------------------------------------------------------------
# 🔟 Save Artifacts
# ---------------------------------------------------------------
model_path = os.path.join(OUTPUT_DIR, "hydroguard_model.h5")
model.save(model_path)

config = {
    "model_name": "HydroGuard_LSTM_v2",
    "features": feature_cols,
    "target": target_col,
    "timesteps": TIMESTEPS,
    "optimizer": "adam",
    "loss": "mse",
    "metrics": ["mae"],
    "note": "Handles string columns via LabelEncoder"
}

config_path = os.path.join(OUTPUT_DIR, "config.yaml")
with open(config_path, "w") as f:
    yaml.dump(config, f)

results = {
    "mse": float(mse),
    "r2_score": float(r2),
    "epochs_trained": len(history.history["loss"]),
    "best_val_loss": float(np.min(history.history["val_loss"]))
}

json_path = os.path.join(OUTPUT_DIR, "results.json")
with open(json_path, "w") as f:
    json.dump(results, f, indent=4)

print(f"[INFO] ✅ Artifacts saved in {OUTPUT_DIR}")

# ---------------------------------------------------------------
# 📊 Visualization
# ---------------------------------------------------------------
plt.figure(figsize=(7,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Training & Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'training_loss_graph.png'))
plt.show()

plt.figure(figsize=(7,4))
plt.plot(y_test[:100], label='Actual', marker='o')
plt.plot(y_pred[:100], label='Predicted', marker='x')
plt.title('Prediction vs Actual (Sample)')
plt.xlabel('Samples')
plt.ylabel('Leak/Burst Index')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'prediction_vs_actual.png'))
plt.show()

print("\n✅ All files successfully generated in:")
print(OUTPUT_DIR)
```
