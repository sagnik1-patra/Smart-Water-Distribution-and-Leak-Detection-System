```python
import os
import json
import yaml
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import pickle
import tensorflow as tf

# ---------------------------------------------------------------
# 1️⃣ Configuration
# ---------------------------------------------------------------
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Water Distribution and Leak Detection System\archive\water_leak_detection_1000_rows.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Water Distribution and Leak Detection System\archive"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ---------------------------------------------------------------
# 2️⃣ Load Dataset
# ---------------------------------------------------------------
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA_PATH)

print("[INFO] Shape:", df.shape)
print("[INFO] Columns:", df.columns.tolist())

df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
df = df.dropna().reset_index(drop=True)

# ---------------------------------------------------------------
# 3️⃣ Encode Non-Numeric Columns
# ---------------------------------------------------------------
encoders = {}
for col in df.columns:
    if df[col].dtype == "object":
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))
        encoders[col] = le
        print(f"[ENCODE] Column '{col}' encoded ({len(le.classes_)} unique values).")

# ---------------------------------------------------------------
# 4️⃣ Feature Selection
# ---------------------------------------------------------------
feature_cols = [col for col in df.columns if col not in ["timestamp", "leak_status", "burst_status"]]
target_col = "burst_status" if "burst_status" in df.columns else feature_cols[-1]

print(f"[INFO] Using {len(feature_cols)} features → {feature_cols}")
print(f"[INFO] Target column → {target_col}")

X = df[feature_cols].values
y = df[target_col].values

# ---------------------------------------------------------------
# 5️⃣ Correlation Heatmap
# ---------------------------------------------------------------
plt.figure(figsize=(8,6))
sns.heatmap(df[feature_cols + [target_col]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hydroguard_heatmap.png"))
plt.show()

# ---------------------------------------------------------------
# 6️⃣ Scale Data
# ---------------------------------------------------------------
scaler_x = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)

with open(os.path.join(OUTPUT_DIR, "scaler.pkl"), "wb") as f:
    pickle.dump(scaler_x, f)
with open(os.path.join(OUTPUT_DIR, "encoders.pkl"), "wb") as f:
    pickle.dump(encoders, f)

# ---------------------------------------------------------------
# 7️⃣ Prepare Sequential Data for LSTM
# ---------------------------------------------------------------
def create_sequences(X, y, timesteps=5):
    Xs, ys = [], []
    for i in range(len(X) - timesteps):
        Xs.append(X[i:(i + timesteps)])
        ys.append(y[i + timesteps])
    return np.array(Xs), np.array(ys)

TIMESTEPS = 5
X_seq, y_seq = create_sequences(X_scaled, y, TIMESTEPS)

split = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]

print(f"[INFO] Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# ---------------------------------------------------------------
# 8️⃣ Build & Train Model
# ---------------------------------------------------------------
model = Sequential([
    LSTM(64, input_shape=(TIMESTEPS, X_train.shape[2]), return_sequences=False),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=16,
    callbacks=[es],
    verbose=1
)

# ---------------------------------------------------------------
# 9️⃣ Evaluate
# ---------------------------------------------------------------
y_pred = model.predict(X_test).flatten()

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"[RESULT] RMSE: {rmse:.4f}, MAE: {mae:.4f}, R²: {r2:.4f}")

# ---------------------------------------------------------------
# 🔟 Save Model & Config
# ---------------------------------------------------------------
model.save(os.path.join(OUTPUT_DIR, "hydroguard_model.h5"))

config = {
    "model_name": "HydroGuard_LSTM_v3",
    "features": feature_cols,
    "target": target_col,
    "timesteps": TIMESTEPS,
    "optimizer": "adam",
    "loss": "mse",
    "metrics": ["mae"]
}
with open(os.path.join(OUTPUT_DIR, "config.yaml"), "w") as f:
    yaml.dump(config, f)

results = {
    "RMSE": float(rmse),
    "MAE": float(mae),
    "R2_Score": float(r2),
    "Epochs_Trained": len(history.history["loss"]),
    "Best_Val_Loss": float(np.min(history.history["val_loss"]))
}
with open(os.path.join(OUTPUT_DIR, "results.json"), "w") as f:
    json.dump(results, f, indent=4)

# ---------------------------------------------------------------
# 🔹 GRAPH 1: Accuracy Graph (Loss + MAE)
# ---------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Training Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'accuracy_graph_loss.png'))
plt.show()

plt.figure(figsize=(8,5))
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Model Accuracy (MAE)')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'accuracy_graph_mae.png'))
plt.show()

# ---------------------------------------------------------------
# 🔹 GRAPH 2: Comparison Graph (Distribution)
# ---------------------------------------------------------------
plt.figure(figsize=(7,5))
sns.kdeplot(y_test, label='Actual', fill=True)
sns.kdeplot(y_pred, label='Predicted', fill=True)
plt.title("Distribution Comparison (Actual vs Predicted)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "comparison_graph.png"))
plt.show()

# ---------------------------------------------------------------
# 🔹 GRAPH 3: Prediction Graph (Line Plot)
# ---------------------------------------------------------------
plt.figure(figsize=(10,5))
plt.plot(y_test[:150], label='Actual', marker='o')
plt.plot(y_pred[:150], label='Predicted', marker='x')
plt.title('Prediction vs Actual (Sample 150)')
plt.xlabel('Sample Index')
plt.ylabel('Leak/Burst Index')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'prediction_graph.png'))
plt.show()

# ---------------------------------------------------------------
# 🔹 GRAPH 4: Result Graph (Metrics Bar)
# ---------------------------------------------------------------
metrics = {'RMSE': rmse, 'MAE': mae, 'R²': r2}
plt.figure(figsize=(6,5))
sns.barplot(x=list(metrics.keys()), y=list(metrics.values()), palette='crest')
plt.title('Model Performance Metrics')
for i, val in enumerate(metrics.values()):
    plt.text(i, val + 0.01, f"{val:.3f}", ha='center', fontweight='bold')
plt.ylabel('Score')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'result_graph.png'))
plt.show()

print("\n✅ All graphs & artifacts successfully generated!")
print("📁 Folder:", OUTPUT_DIR)
print("\nFiles created:")
print(" - hydroguard_model.h5")
print(" - scaler.pkl")
print(" - encoders.pkl")
print(" - config.yaml")
print(" - results.json")
print(" - hydroguard_heatmap.png")
print(" - accuracy_graph_loss.png")
print(" - accuracy_graph_mae.png")
print(" - comparison_graph.png")
print(" - prediction_graph.png")
print(" - result_graph.png")
```
