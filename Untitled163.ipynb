```python
import os, json, yaml, pickle, random
import numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf

# ---------------------------------------------------------------
# 1️⃣ Configuration
# ---------------------------------------------------------------
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Water Distribution and Leak Detection System\archive\water_leak_detection_1000_rows.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Water Distribution and Leak Detection System\archive"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ---------------------------------------------------------------
# 2️⃣ Load & preprocess data
# ---------------------------------------------------------------
df = pd.read_csv(DATA_PATH)
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
df = df.dropna().reset_index(drop=True)

# Encode categorical
encoders = {}
for col in df.columns:
    if df[col].dtype == "object":
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))
        encoders[col] = le

# Feature/target
feature_cols = [c for c in df.columns if c not in ["timestamp","leak_status","burst_status"]]
target_col = "burst_status" if "burst_status" in df.columns else feature_cols[-1]
X, y = df[feature_cols].values, df[target_col].values

# Scale
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

def create_seq(X,y,steps=5):
    Xs,ys=[],[]
    for i in range(len(X)-steps):
        Xs.append(X[i:i+steps])
        ys.append(y[i+steps])
    return np.array(Xs), np.array(ys)

TIMESTEPS=5
X_seq,y_seq=create_seq(X_scaled,y,TIMESTEPS)
split=int(0.8*len(X_seq))
X_train,X_test=X_seq[:split],X_seq[split:]
y_train,y_test=y_seq[:split],y_seq[split:]

# ---------------------------------------------------------------
# 3️⃣ Helper: build & evaluate LSTM
# ---------------------------------------------------------------
def build_eval(params):
    hidden,drop,lr=int(params[0]),params[1],params[2]
    model=Sequential([
        LSTM(hidden,input_shape=(TIMESTEPS,X_train.shape[2])),
        Dropout(drop),
        Dense(32,activation='relu'),
        Dense(1,activation='linear')
    ])
    model.compile(optimizer=Adam(learning_rate=lr),loss='mse')
    es=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
    hist=model.fit(X_train,y_train,validation_data=(X_test,y_test),
                   epochs=40,batch_size=16,verbose=0,callbacks=[es])
    val_loss=np.min(hist.history["val_loss"])
    return val_loss,model,hist

# ---------------------------------------------------------------
# 4️⃣ AIS + Clonal Selection Optimization
# ---------------------------------------------------------------
POP_SIZE=8
GENERATIONS=5
MUT_RATE=0.3
CANDIDATES=[ [random.randint(32,128),
              round(random.uniform(0.1,0.5),2),
              round(random.uniform(0.0005,0.01),4)]
             for _ in range(POP_SIZE)]

best_val,best_model,best_hist=float("inf"),None,None
loss_evolution=[]

print("[INFO] Starting AIS + CSA optimization...\n")

for g in range(GENERATIONS):
    print(f"🧬 Generation {g+1}/{GENERATIONS}")
    scored=[]
    for p in CANDIDATES:
        val,model,hist=build_eval(p)
        scored.append((val,p,model,hist))
        print(f"  Params {p} → ValLoss={val:.4f}")
    scored.sort(key=lambda x:x[0])
    best_val,best_params,best_model,best_hist=scored[0]
    loss_evolution.append(best_val)
    # Clonal selection: replicate top 3 with mutations
    new=[]
    for _,p,_,_ in scored[:3]:
        for _ in range(POP_SIZE//3):
            clone=p.copy()
            if random.random()<MUT_RATE:
                clone[0]=max(32,min(128,clone[0]+random.randint(-8,8)))
            if random.random()<MUT_RATE:
                clone[1]=round(min(0.5,max(0.1,clone[1]+random.uniform(-0.05,0.05))),2)
            if random.random()<MUT_RATE:
                clone[2]=round(min(0.01,max(0.0005,clone[2]*random.uniform(0.8,1.2))),4)
            new.append(clone)
    CANDIDATES=new
    print(f"  Best so far: {best_params} → {best_val:.4f}\n")

print("[INFO] Optimization complete!\nBest Params:",best_params)

# ---------------------------------------------------------------
# 5️⃣ Evaluate best model
# ---------------------------------------------------------------
y_pred=best_model.predict(X_test).flatten()
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
rmse=np.sqrt(mse)
print(f"[RESULT] RMSE={rmse:.4f}  MAE={mae:.4f}  R2={r2:.4f}")

# ---------------------------------------------------------------
# 6️⃣ Save results & config
# ---------------------------------------------------------------
best_model.save(os.path.join(OUTPUT_DIR,"hybrid_hydroguard_model.h5"))
with open(os.path.join(OUTPUT_DIR,"scaler.pkl"),"wb") as f: pickle.dump(scaler,f)
with open(os.path.join(OUTPUT_DIR,"encoders.pkl"),"wb") as f: pickle.dump(encoders,f)

results={"RMSE":float(rmse),"MAE":float(mae),"R2":float(r2),
         "Best_Params":best_params,"Val_Loss_Evolution":loss_evolution}
with open(os.path.join(OUTPUT_DIR,"hybrid_results.json"),"w") as f:
    json.dump(results,f,indent=4)

config={"model":"Hybrid_AIS_CSA_LSTM","timesteps":TIMESTEPS,
        "optimizer":"adam","feature_count":len(feature_cols)}
with open(os.path.join(OUTPUT_DIR,"hybrid_config.yaml"),"w") as f:
    yaml.dump(config,f)

# ---------------------------------------------------------------
# 7️⃣ Graphs
# ---------------------------------------------------------------

# 📊 1. Hybrid accuracy graph (val loss vs generations)
plt.figure(figsize=(8,5))
plt.plot(loss_evolution,marker='o')
plt.title("Hybrid AIS+CSA Optimization Loss")
plt.xlabel("Generation")
plt.ylabel("Best Validation Loss")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR,"hybrid_accuracy_graph_loss.png"))
plt.show()

# 📊 2. Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[feature_cols+[target_col]].corr(),annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Feature Correlation Heatmap (Hybrid)")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR,"hybrid_heatmap.png"))
plt.show()

# 📊 3. Comparison graph
plt.figure(figsize=(7,5))
sns.kdeplot(y_test,label="Actual",fill=True)
sns.kdeplot(y_pred,label="Predicted",fill=True)
plt.title("Distribution Comparison (Hybrid)")
plt.xlabel("Value"); plt.ylabel("Density")
plt.legend(); plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR,"hybrid_comparison_graph.png"))
plt.show()

# 📊 4. Prediction graph
plt.figure(figsize=(10,5))
plt.plot(y_test[:150],label="Actual",marker='o')
plt.plot(y_pred[:150],label="Predicted",marker='x')
plt.title("Prediction vs Actual (Hybrid)")
plt.xlabel("Sample Index"); plt.ylabel("Leak/Burst Index")
plt.legend(); plt.grid(True); plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR,"hybrid_prediction_graph.png"))
plt.show()

# 📊 5. Result graph
metrics={"RMSE":rmse,"MAE":mae,"R²":r2}
plt.figure(figsize=(6,5))
sns.barplot(x=list(metrics.keys()),y=list(metrics.values()),palette="crest")
plt.title("Hybrid Model Performance Metrics")
for i,v in enumerate(metrics.values()):
    plt.text(i,v+0.01,f"{v:.3f}",ha="center",fontweight="bold")
plt.ylabel("Score"); plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR,"hybrid_result_graph.png"))
plt.show()

print("\n✅ Hybrid AIS+CSA results generated successfully!")
print("📁 Folder:", OUTPUT_DIR)
```
